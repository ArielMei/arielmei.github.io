# Get data from website and convert to dataframe

````python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
````

````python
url=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(url,'lxml')
#print(soup.prettify())    # print the full text of the website
````

````python
table = soup.find('table',{'class':'wikitable sortable'})    # wikitable sortable is the desired table name
table
````

````python
ths = table.find_all('th')                 #  Columns' name are seperated by th
headings = [th.text.strip() for th in ths]
headings
````

````python
lists=[]                   #Get each row and append to a list
for row in table.findAll("tr"):
    cells = row.findAll("td")
    items = [th.text.strip() for th in cells]
    lists.append(items)
````

````python
df=pd.DataFrame(lists[1:])     #Convert the list to dataframe
df.columns=headings
df.head()
````

````python
import pandas as pd
import xml.etree.cElementTree as et
 
parsedXML = et.parse("books.xml" )
dfcols = ['id', 'author', 'title', 'genre','price','publish_data','description']
df_xml = pd.DataFrame(columns=dfcols)
````

````python
def getvalueofnode(node):
    """ return node text or None """
    return node.text if node is not None else None

for node in parsedXML.getroot():
    id = node.attrib.get('id')
    author = node.find('author')
    title = node.find('title')
    genre = node.find('genre')
    price = node.find('price')
    publish_date = node.find('publish_date')
    description = node.find('description')
    
    df_xml = df_xml.append(
            pd.Series([id, 
                       getvalueofnode(author), 
                       getvalueofnode(title),
                       getvalueofnode(genre),
                       getvalueofnode(price),
                       getvalueofnode(publish_date),
                       getvalueofnode(description)], index=dfcols),
            ignore_index=True)
````
